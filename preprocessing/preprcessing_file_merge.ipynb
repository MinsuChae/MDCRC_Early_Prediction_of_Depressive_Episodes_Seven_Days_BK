```python
import pandas as pd

df_heart_rate = pd.read_csv("feature_heart_rate.csv")
df_hrv = pd.read_csv("HRV.csv")
df_hrv.rename(columns={'date':'timestamp'}, inplace=True)
df_light = pd.read_csv("feature_light.csv")
df_sleep = pd.read_csv("feature_sleep_DLMO.csv")
df_step = pd.read_csv("feature_steps.csv")
df_cr = pd.read_csv("feature_circadian_rhythm.csv")
df = pd.read_csv("label_MDCRC.csv")
df.drop('ep count', axis=1, inplace=True)
```

```python
df_tmp = pd.merge(df_heart_rate, df_hrv, on=["patient_id","timestamp"], how='outer')
df_tmp = pd.merge(df_tmp, df_light, on=["patient_id","timestamp"], how='outer')
df_tmp = pd.merge(df_tmp, df_sleep, on=["patient_id","timestamp"], how='outer')
df_tmp = pd.merge(df_tmp, df_step, on=["patient_id","timestamp"], how='outer')
df_tmp = pd.merge(df_tmp, df_cr, on=["patient_id","timestamp"], how='outer')
df = pd.merge(df_tmp, df, on=["patient_id","timestamp"], how='right')
df = df.sort_values(["patient_id","timestamp"])
```
